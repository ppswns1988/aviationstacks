## Conclusiones

El proyecto proporcionó información valiosa sobre los patrones de retraso de vuelos y los factores que los influyen. En términos generales, la mayoría de las aerolíneas presentaron una media de retraso entre 30 y 37 minutos. Es importante destacar que los retrasos en los vuelos afectan tanto a vuelos comerciales como de carga.

En el período de análisis, encontramos que las aerolíneas extranjeras tienen una mayor presencia en el mercado, con 6 empresas representadas, seguidas de 3 aerolíneas de origen español.

El país que posee mayor número de aerolineas es China. Otro aspecto importante, el país con más flota sigue siendo Estados Unidos (4000 aviones), y tomando como referencia ese apartado, con menos florta de aviones esta Zambia con un solo avión. La flota mas longeva la lideran Colombia, South Africa y Venezuela.

EL mayor propietario de aviones en todo el mundo, es de los Estados unidos, Wells Fargo Bank.
