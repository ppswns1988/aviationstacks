```python
import pandas as pd
import numpy as np
import ast
import datetime
```

```python
df_dep_historical_12 = pd.read_csv("../Proyecto/df_dep_historical_12.csv")
df_dep_historical_12
```

```python
#Eliminar columnas no necesarias
df_dep_historical_12 = df_dep_historical_12.drop('flight_status', axis=1)
df_dep_historical_12 = df_dep_historical_12.drop('aircraft', axis=1)
df_dep_historical_12 = df_dep_historical_12.drop('live', axis=1)
```

```python
#Cambio a Datetime
df_dep_historical_12["flight_date"] = pd.to_datetime(df_dep_historical_12["flight_date"])
```

```python
df_dep_historical_12["flight_date"] = df_dep_historical_12["flight_date"].dt.strftime("%y/%m/%d")
```

```python
#Columna departure:
```

```python
#Moficacion a string
df_dep_historical_12['departure'] = df_dep_historical_12['departure'].astype(str)
```

```python
#columna departure de una cadena de texto a diccionario de Python utilizando ast.literal_eval.
#nuevas columnas de diccionario usando apply(pd.Series)
dict_departure= df_dep_historical_12['departure'].apply(ast.literal_eval).apply(pd.Series) 
```

```python
#Key a columnas
keys_to_colum_departure = ['airport', 'timezone', 'iata', 'icao', 'terminal', 'gate', 'delay']
```

```python
#Filtrar columnas a usar del diccionario
departure_filter = dict_departure[keys_to_colum_departure]
```

```python
# Concatenar las nuevas columnas al df
df_dep_historical_12 = pd.concat([df_dep_historical_12.drop(columns=['departure']), departure_filter], axis=1)
```

```python
df_dep_historical_12
```

```python
#Columna arrival
```

```python
#Moficacion a string
df_dep_historical_12['arrival'] = df_dep_historical_12['arrival'].astype(str)
```

```python
#columna arrival de una cadena de texto a diccionario de Python utilizando ast.literal_eval.
#nuevas columnas de diccionario usando apply(pd.Series)
dict_arrival= df_dep_historical_12['arrival'].apply(ast.literal_eval).apply(pd.Series) 
```

```python
#Key a columnas
keys_to_colum_arrival = ['airport', 'timezone', 'iata', 'icao', 'terminal', 'gate', 'baggage', 'delay']
```

```python
#Filtrar columnas a usar del diccionario
arrival_filter = dict_arrival[keys_to_colum_arrival]
```

```python
# Concatenar las nuevas columnas al df
df_dep_historical_12 = pd.concat([df_dep_historical_12.drop(columns=['arrival']), arrival_filter], axis=1)
```

```python
df_dep_historical_12
```

```python
#Columna airline
```

```python
#Moficacion a string
df_dep_historical_12['airline'] = df_dep_historical_12['airline'].astype(str)
```

```python
#columna airline de una cadena de texto a diccionario de Python utilizando ast.literal_eval.
#nuevas columnas de diccionario usando apply(pd.Series)
dict_airline= df_dep_historical_12['airline'].apply(ast.literal_eval).apply(pd.Series) 
```

```python
#Key a columnas
keys_to_colum_airline = ['name', 'iata','icao']
```

```python
#Filtrar columnas a usar del diccionario
airline_filter = dict_airline[keys_to_colum_airline]
```

```python
# Concatenar las nuevas columnas al df
df_dep_historical_12 = pd.concat([df_dep_historical_12.drop(columns=['airline']), airline_filter], axis=1)
```

```python
df_dep_historical_12
```

```python
#Columna flight
```

```python
#Moficacion a string
df_dep_historical_12['flight'] = df_dep_historical_12['flight'].astype(str)
```

```python
#columna airline de una cadena de texto a diccionario de Python utilizando ast.literal_eval.
#nuevas columnas de diccionario usando apply(pd.Series)
dict_flight= df_dep_historical_12['flight'].apply(ast.literal_eval).apply(pd.Series)
```

```python
#Key a columnas
keys_to_colum_flight = ['airline_name', 'airline_iata', 'airline_icao', 'flight_number', 'flight_iata', 'flight_icao']
```

```python
#Entrar al diccionario andiado 'codeshared' para extraer los kye de columnas
def extraer_anidado(filas, keys , key_anidada):
    resultado ={}
    dict_nest = filas.get(key_anidada,{})
    if isinstance (dict_nest, dict):
        for key in keys:
            resultado[key] = dict_nest.get(key,None)
    return pd.Series(resultado)
```

```python
#Extraer las columnas del diccionario anidado con una funcion
codeshared_filter = dict_flight.apply(lambda filas: extraer_anidado(filas, keys_to_colum_flight,'codesshared'), axis = 1)
```

```python
# Concatenar las nuevas columnas al df
df_dep_historical_12 = pd.concat([df_dep_historical_12.drop(columns=['flight']), codeshared_filter], axis=1)
```

```python
df_dep_historical_12
```

```python
#Los sigiuentes codigos permiten encontrar los indices de las columnas cuyos nombres estan duplicados y renombrar las columnas 
#a efectos de hacer drop a la columna repetida que no representa importacia al analisis.
```

```python
#Renombrar la columna 'airport'
airport_columns = [i for i, col in enumerate(df_dep_historical_12.columns) if col == 'airport']
if len(airport_columns) > 1:
    df_dep_historical_12.columns.values[airport_columns[1]] = 'arr_airport'
```

```python
#Renombrar la columna 'iata'
iata_columns = [i for i, col in enumerate(df_dep_historical_12.columns) if col == 'iata']
if len(iata_columns) > 2:
    df_dep_historical_12.columns.values[iata_columns[2]] = 'iata_airplane'
```

```python
#Renombrar la columna 'timezone'
timezone_columns = [i for i, col in enumerate(df_dep_historical_12.columns) if col == 'timezone']
if len(timezone_columns) > 1:
    df_dep_historical_12.columns.values[timezone_columns[1]] = 'arr_timezone'
```

```python
#Renombrar la columna 'delay'
delay_columns = [i for i, col in enumerate(df_dep_historical_12.columns) if col == 'delay']
if len(delay_columns) > 1:
    df_dep_historical_12.columns.values[delay_columns[1]] = 'arr_delay'
```

```python
#Renombrar la columna 'terminal'
terminal_columns = [i for i, col in enumerate(df_dep_historical_12.columns) if col == 'terminal']
if len(terminal_columns) > 1:
    df_dep_historical_12.columns.values[terminal_columns[1]] = 'arr_terminal'
```

```python
#Renombrar la columna 'gate'
gate_columns = [i for i, col in enumerate(df_dep_historical_12.columns) if col == 'gate']
if len(gate_columns) > 1:
    df_dep_historical_12.columns.values[gate_columns[1]] = 'gate_iata'
```

```python
#Renombrar la columna 'iata'
iata_columns = [i for i, col in enumerate(df_dep_historical_12.columns) if col == 'iata']
if len(iata_columns) > 1:
    df_dep_historical_12.columns.values[iata_columns[1]] = 'arr_iata'
```

```python
#Renombrar columna 'name para dar mas sentido por nombre de aerolinea
df_dep_historical_12.rename(columns={'name': 'airline_names'}, inplace=True)
```

```python
#Eliminar columna 'icao'
df_dep_historical_12.drop(columns=['icao'], axis=1, inplace=True)
```

```python
#Eliminar columna 'arr_delay'
df_dep_historical_12.drop(columns=['arr_delay'], axis=1, inplace=True)
```

```python
#Eliminar columna 'gate_iata'
df_dep_historical_12.drop(columns=['gate_iata'], axis=1, inplace=True)
```

```python
#Eliminamos las primeras columnas no representativas para los datos
df_dep_historical_12.drop(columns=['airline_name','airline_iata','airline_icao','flight_number','flight_iata','flight_icao', 'arr_terminal'], inplace = True)
```

```python
#Calculo de mean para uasrla como fillna en columna delay
mean_delay = np.round(df_dep_historical_12['delay'].mean(), 1)
```

```python
#Asignar valor de la media de 'delay' la media con round(2)
df_dep_historical_12.fillna({'delay': mean_delay}, inplace=True)
```

```python
#Asignar valores a gate con fillna
df_dep_historical_12['gate'] = df_dep_historical_12['gate'].fillna('X99')
```

```python
#Convertir los valores None a '0'en columna baggega
df_dep_historical_12['baggage'] = df_dep_historical_12['baggage'].replace({None: 0})
```

```python
#Asignar valores 'o' usando pd.to.numeric y errors ='coerce' para convertir a numericos y no numericos en 'NaN'
df_dep_historical_12['baggage'] = pd.to_numeric(df_dep_historical_12['baggage'], errors='coerce').fillna(0)
```

```python
#Cambiar tipo de dato a 'int'
df_dep_historical_12['baggage'] = df_dep_historical_12['baggage'].astype(int)
```

```python
#Calcular mean de 'baggage'
mean_baggage = np.round(df_dep_historical_12['baggage'].mean(), 1)
```

```python
#Asignar mean a valores numericos o NaN
df_dep_historical_12.fillna({'baggage': mean_baggage}, inplace=True)
```

```python
df_dep_historical_12
```

```python
df_dep_historical_12.to_csv('df_dep_hist_12_clean.csv', index=False)
```

```python
#Key a columnas
keys_to_colum_flight = ["airline_name", "airline_iata", "flight_number", "flight_iata"]
```

```python
#Entrar al diccionario andiado 'codeshared' para extraer los kye de columnas
def extraer_anidado(filas, keys , key_anidada):
    resultado ={}
    dict_nest = filas.get(key_anidada,{})
    if isinstance (dict_nest, dict):
        for key in keys:
            resultado[key] = dict_nest.get(key,None)
    return pd.Series(resultado)
```

```python
#Extraer las columnas del diccionario anidado con una funcion
codeshared_filter = dict_flight.apply(lambda filas: extraer_anidado(filas, keys_to_colum_flight,'codesshared'), axis = 1)
```

```python
# Concatenar las nuevas columnas al df
df_arr_historical_14 = pd.concat([df_arr_historical_14.drop(columns=["flight"]), codeshared_filter], axis=1)
```
