```python
import numpy as np
import pandas as pd
import tqdm
import requests
import os
import json
from pprint import pprint
```

**1. DF_TAXES**

```python
#DF_TAXES
df_taxes = pd.read_csv("../Proyecto/DF_LIMPIOS/taxes_limpio.csv")

df_taxes = df_taxes.replace({np.nan : None})

df_taxes
```

***SUBIDA DF_TAXES (no ejecutar! se duplican los datos)***

```python
#DF_TAXES
# Define the base ID and table name
base_id = 'appniGTOSp2oeL7IH'
table_name = "tblmobYwOeq2zmD6Y"
airtable_base_url = "https://api.airtable.com/v0"

# Define the API token
token = 'pat0OxeaZtOb3D3xh.2badecd5b26991ad4831181e0b85e4cb1449bf9352b968e2cf282ddf90365568'

endpoint = f"https://api.airtable.com/v0/{base_id}/{table_name}"

headers = {
    "Authorization": f"Bearer {token}",
    "Content-Type": "application/json"
}
```

```python
#DF_TAXES
total_rows = len(df_taxes)

current = 0

while current < total_rows:
    until = current + 10
    print(f"Subiendo filas desde {current} hasta {until}")

    # Subir los 10 datos de cada iteración
    records = {"records"  : [],
               "typecast" : True}

    for i in range(current, until):
        if i < total_rows:
            fila = {"fields" : df_taxes.iloc[i].to_dict()}
            records["records"].append(fila)

    response = requests.post(url = endpoint, json = records, headers = headers)

    # aumentar en 10 el contador
    current += 10
```

***DESCARGA DF_TAXES***

```python
%%time
#DF_TAXES
params = {}

endpoint = f"{airtable_base_url}/{base_id}/{table_name}"

datos = list()

while params.get("offset") != None or len(datos) == 0:
    
    response = requests.get(url = endpoint, headers = headers, params = params)
    
    print(f"response: {response.status_code}")
    print(f"endpoint: {response.url}")
    
    data = response.json()
    
    offset = data.get("offset")
    
    params["offset"] = offset
    
    datos.extend(data["records"])
```

```python
df_taxes = pd.json_normalize(datos)

df_taxes.to_csv("taxes.csv")
```
