```python
import requests
import pandas as pd
import tqdm
```

**Airports**

```python
%%time

access_key = "755bf3133a02b15cd9b5ce2998c023b5"
url = f"http://api.aviationstack.com/v1/airports?access_key={access_key}"
params = {"limit": 100, "offset": 0}

data = []
for i in tqdm.tqdm(range(70)):
    response = requests.get(url, params=params)
    response_json = response.json()
    if 'data' in response_json:
        data.extend(response_json["data"])
    else:
        print("Error:", response_json)
    params["offset"] += 100

df = pd.DataFrame(data)
df_airports = pd.DataFrame(data)
df_airports
```
